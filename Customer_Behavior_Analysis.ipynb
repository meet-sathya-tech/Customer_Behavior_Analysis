```python
import pandas as pd
df = pd.read_csv('customer_shopping_behavior.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe(include = "all")# by default it'll give only numeric so we comes include="all"
```

```python
df.isnull().sum()#checking missing values
```

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x:x.fillna(x.median()))
#using median over mean is good and doing that to for every each category , not like for alll the category which will ends up by wrong
```

```python
df.isnull().sum()#checking missing values
```

```python
df.columns = df.columns.str.lower()#capital case to lower case
df.columns = df.columns.str.replace(' ','_')# empty space btw words to '_'
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
#create a new column age_group
labels = ['Young Adult','Adult','Middle Age','Senior']
df['age_group'] = pd.qcut(df['age'],q = 4,labels = labels)
```

```python
df[['age','age_group']]
```

```python
#purchase_frequency_days

frequency_mapping = {
    'Fortnightly':14,
    'Weekly':7,
    'Monthly':30,
    'Quarterly':90,
    'Bi-Weekly':14,
    'Annually':365,
    'Every 3 Months':90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```
